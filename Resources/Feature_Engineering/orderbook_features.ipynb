# filter df to within market hours

```python
# Convert your string timestamps to timezone-aware Timestamp objects
start_time = pd.to_datetime("2023-10-25 13:30:00.000000000").tz_localize('UTC')
end_time = pd.to_datetime("2023-10-25 20:00:00.000000000").tz_localize('UTC')

# Now filter using query by directly using these Timestamp objects
df_filtered = df.query('@start_time <= ts_recv <= @end_time')
```

```python
bid_price_columns = [col for col in df_filtered.columns if 'bid_px_' in col]
bid_size_columns = [col for col in df_filtered.columns if 'bid_sz_' in col]
ask_price_columns = [col for col in df_filtered.columns if 'ask_px_' in col]
ask_size_columns = [col for col in df_filtered.columns if 'ask_sz_' in col]
heatmap_data = pd.DataFrame()

for i in range(len(bid_price_columns)):
    temp_df = df_filtered[['ts_event', bid_price_columns[i], bid_size_columns[i]]].copy()
    temp_df.rename(columns={bid_price_columns[i]: 'price', bid_size_columns[i]: 'volume'}, inplace=True)
    temp_df['volume'] = np.log(temp_df['volume'])
    heatmap_data = pd.concat([heatmap_data, temp_df])
for i in range(len(ask_price_columns)):
    temp_df = df_filtered[['ts_event', ask_price_columns[i], ask_size_columns[i]]].copy()
    temp_df.rename(columns={ask_price_columns[i]: 'price', ask_size_columns[i]: 'volume'}, inplace=True)
    #temp_df['volume'] = -temp_df['volume']
    temp_df['volume'] = -np.log(temp_df['volume'])
    heatmap_data = pd.concat([heatmap_data, temp_df])
```

```python
heatmap_data = heatmap_data.groupby(['ts_event', 'price']).sum().reset_index()
```

```python
heatmap_data_pivot = heatmap_data.pivot(index = 'ts_event', columns = 'price', values = 'volume')
```

```python
heatmap_data_pivot = heatmap_data_pivot.iloc[:, ::-1]
```

```python
values = heatmap_data_pivot.values.flatten()
values = values[np.logical_not(np.isnan(values))]
ask_values = values[values < 0]
bid_values = values[values > 0]
```

```python
plt.figure(figsize=(12, 6))
plt.hist(np.clip(-ask_values,0,20), bins=200, alpha=0.5, label='ask', color='r')
plt.hist(np.clip(bid_values,0,20), bins=200, alpha=0.5, label='bid', color='b')
plt.title('Log Volume distribution')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Create the heatmap
plt.figure(figsize=(20, 10))
plot_map_df = heatmap_data_pivot[100000:110000].dropna(axis=1, how='all').T
sns.heatmap(plot_map_df, cmap='RdYlGn', center=0) # Use 'RdYlGn' colormap for green and red colors
plt.figtext(0.5, 0.005,'Order Book Volume Over Time (1000 events during normal market)', ha="center", fontsize=10, bbox={"facecolor":"orange", "alpha":0.5, "pad":5})
plt.title('Order Book Volume Over Time')
plt.xlabel('Time', labelpad=20, loc='right')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```
